## **Calories Burnt Prediction**

>**Importing Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from xgboost import XGBRegressor
from sklearn import metrics
```

>**Data Collection**

```python
# loading the data from csv file to a DataFrame

calories = pd.read_csv('/content/calories.csv')
exercise_data = pd.read_csv('/content/exercise.csv')
```

```python
calories.head()
exercise_data.head()
```

```python
# Combining the two Dataframes
calories_data = pd.concat([exercise_data, calories['Calories']], axis=1)
calories_data.head()
```

```python
#number of rows and columns
calories_data.shape
```

```python
calories_data.info()
```

```python
# checking for missing values
calories_data.isnull().sum()
```

```python
# statistics
calories_data.describe()
```

>**Data Visualization**

```python
sns.set()
```

```python
# finding gender ratio using count plot
sns.countplot(calories_data['Gender'])
```

```python
# finding the distribution of age
sns.distplot(calories_data['Age'])
```

```python
# finding the distribution of height
sns.distplot(calories_data['Height'])
```

```python
# finding the distribution of weight
sns.distplot(calories_data['Weight'])
```

```python
#dataset with all numeric values

label_encoder = LabelEncoder()
calories_data2 = calories_data.copy()
calories_data2['gender_encoded'] = label_encoder.fit_transform(calories_data2['Gender'])
calories_data2.drop('Gender', axis=1, inplace=True)
calories_data2
```

>**Pre-Processing**

```python
# Check Correlation in dataset

correlation = calories_data2.corr()
```

```python
# correlation visualisation using heatmap

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
#Converting the text data to numerical values

calories_data.replace({"Gender":{'male':0,'female':1}}, inplace=True)
calories_data.head()
```

```python
#Separating features and Target

X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']

print(X)
```

>**Train Test Split**

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

>**Model Training** - XGBoost Regressor



```python
model = XGBRegressor()
```

```python
# training the model with X_train
model.fit(X_train, Y_train)
```

```python
test_data_prediction = model.predict(X_test)
print(test_data_prediction)
```

Mean Absolute Error

```python
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print("Mean Absolute Error = ", mae)
```

```python
#First 'n' no. of predicted Data
Y_test2 = Y_test[:100]
test_data_prediction2 = test_data_prediction[:100]

#Comparison between Actual and Predicted Calories
plt.figure(figsize=(8, 6))
plt.scatter(Y_test2, test_data_prediction2, color='blue', label='Actual vs Predicted Calories')

plt.plot(np.unique(Y_test2), np.poly1d(np.polyfit(Y_test2, test_data_prediction2, 1))(np.unique(Y_test2)), color='red')

plt.title('Actual vs Predicted Calories')
plt.xlabel('Actual Calories')
plt.ylabel('Predicted Calories')
plt.legend()

plt.grid(True)
plt.show()
```
